```python
from selenium import webdriver 
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import csv

def retrieve_details(login_document):
    with open(login_document, newline='') as f:
        reader = csv.reader(f)
        credentials = list(reader)
    username = credentials[1][1]
    password = credentials[2][1]
    networkUser = credentials[3][1]
    networkPass = credentials[4][1]
    
    return username, password, networkUser, networkPass
    
def ntulearn_login(browser,username,password):
    browser.get(('https://ntulearn.ntu.edu.sg'))
    
    WebDriverWait(browser,10).until(EC.visibility_of_element_located((By.ID,'userNameInput')))
    browser.find_element_by_id('userNameInput').send_keys(username)

    password = browser.find_element_by_id('passwordInput').send_keys(password)

    browser.find_element_by_id('submitButton').click()

def close_popups(browser):
    WebDriverWait(browser,10).until(EC.visibility_of_element_located((By.CLASS_NAME,'button-1')))
    WebDriverWait(browser,10).until(EC.element_to_be_clickable((By.CLASS_NAME,'button-1'))).click()

    WebDriverWait(browser,10).until(EC.visibility_of_element_located((By.ID,'quriobotPopupClose')))
    WebDriverWait(browser,10).until(EC.element_to_be_clickable((By.ID,'quriobotPopupClose'))).click()

def open_temp_recorder(browser,networkUser,networkPass):
    #xpathlink = '//a[@href="https://sso.wis.ntu.edu.sg/webexe88/owa/sso_login1.asp?t=1&amp;p2=https://wis.ntu.edu.sg/pls/webexe/str_stud.BRANCH_STUD"]' 
    #WebDriverWait(browser,10).until(EC.element_to_be_clickable((By.XPATH,xpathlink))).click()
    #WebDriverWait(browser,10).until(EC.visibility_of_element_located((By.PARTIAL_LINK_TEXT,"Student Temperature")))
    WebDriverWait(browser,10).until(EC.element_to_be_clickable((By.PARTIAL_LINK_TEXT,"Student Temperature"))).click()
    WebDriverWait(browser,10).until(EC.visibility_of_element_located((By.NAME,"UserName"))).send_keys(networkUser)

    okButton = browser.find_element_by_name("bOption")
    okButton.click()

    WebDriverWait(browser,10).until(EC.visibility_of_element_located((By.NAME,"PIN"))).send_keys(networkPass)

    browser.find_element_by_xpath("//input[@type='submit']").click()

    WebDriverWait(browser,10).until(EC.visibility_of_element_located((By.XPATH,"//input[@value='Confirm']")))
    WebDriverWait(browser,10).until(EC.element_to_be_clickable((By.XPATH,"//input[@value='Confirm']"))).click()
    
    WebDriverWait(browser,10).until(EC.element_to_be_clickable((By.XPATH,"//input[@value='View Previous Readings']"))).click()
    
    time.sleep(4)
    
    browser.quit()

def submit_temp(browser):
    retrieve_details("login_details.csv")
    ntulearn_login(browser,retrieve_details("login_details.csv")[0],retrieve_details("login_details.csv")[1])
    close_popups(browser)
    open_temp_recorder(browser,retrieve_details("login_details.csv")[2],retrieve_details("login_details.csv")[3])

browser = webdriver.Chrome("chromedriver")
submit_temp(browser)
```
